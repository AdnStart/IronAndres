```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression # OLS algorithm
from sklearn.linear_model import Ridge # Lasso algorithm
from sklearn.neighbors import KNeighborsRegressor
```

```python
#creating dataframe
df = pd.read_csv("Data_Marketing_Customer_Analysis_Round2.csv")
df.drop(columns=["Unnamed: 0"], inplace = True)
df
```

```python
df = df.dropna(axis=0)
```

```python
numerics_df = df._get_numeric_data().drop(columns=["Total Claim Amount"])
numerics_df
```

```python
X= numerics_df
y= df["Total Claim Amount"]
```

```python
y
```

# OLS

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
ols=LinearRegression()    # model

ols.fit(X_train.to_numpy(), y_train)   # model train

ols_yhat=ols.predict(X_test.to_numpy())   # model prediction
y_pred_train=ols.predict(X_train.to_numpy())
```

```python
#R2
r2= ols.score(X_test.to_numpy(), y_test)
print(r2)
```

```python
#adjusted rsquare
N=X_train.shape[0]
p=7
x = (1-r2)
y = (N-1) / (N-p-1)
adj_rsquared = (1 - (x * y))
print("Adjusted-R2 : " , adj_rsquared)
```
